```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../parser/vacancies.csv")
```

**Top technologies to learn based on Djinni**

```python
import heapq

technology_count_mentions = {}

for row in df["technologies_mentioned"].str.split(","):
    for item in row:
        if item not in technology_count_mentions:
            technology_count_mentions[item] = 0
        technology_count_mentions[item] += 1

top20_technologies = heapq.nlargest(20, technology_count_mentions.keys(), key=technology_count_mentions.get)
```

```python
top20_technologies_dict = {tech: technology_count_mentions[tech]
                           for tech in top20_technologies}

top20_technologies_dict = sorted(top20_technologies_dict.items(), key=lambda x: x[1])

labels, values = zip(*top20_technologies_dict)

plt.figure(figsize=(10, 6))
plt.barh(labels, values, color="skyblue")
plt.xlabel("Number of Mentions")
plt.ylabel("Technology")
plt.title("Technology Mentions In Djinni Vacancies")
plt.tight_layout()

plt.show()
```

```python
english_level_counts = df["english_level"].value_counts()

plt.figure(figsize=(8, 8))
pie = english_level_counts.plot.pie(autopct="%1.1f%%", startangle=270)
pie.set_ylabel("") 
plt.title("English Level Required For Vacancy")

plt.show()
```

```python
experience_level_count = df["experience_years"].value_counts()

plt.figure(figsize=(8, 8))
pie = experience_level_count.plot.pie(autopct="%1.1f%%", startangle=270)
pie.set_ylabel("") 
plt.title("Number of Vacancies by Experience Level")

plt.show()
```
